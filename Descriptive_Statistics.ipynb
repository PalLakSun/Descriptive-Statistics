1.Q1: Understanding Central Tendency (Easy)
A bakery tracks the daily sales of muffins (in dozens) over a week: [10, 12, 11, 15, 14, 13, 12].
What is the most representative value of their weekly sales, and why?

Ans-Given Sales Data:[10, 12, 11, 15, 14, 13, 12]

Mean=Average=(10+12+11+15+14+13+127)/7=87/7=12.43

Median=Middle value when the data is ordered

AO- [10, 11, 12, 12, 13, 14, 15]

Middle value is 12 since the data points are odd

Mode=Most frequently value.Here, 12 appears twice,while all other values appear once so mode is 12

Mode: 12 (most frequent value)

Median: 12 (middle value)

Mean: 12.43 (average value)

Since the mean is very close to the median and mode, we can say that 12 (both the mode and median) is the most representative value for the weekly sales. It reflects the typical sales day most accurately.

A finance student compares the daily price fluctuations of two cryptocurrencies.

Coin A: standard deviation = $30

Coin B: standard deviation = $120
 Which coin is riskier to invest in, and why?

 Ans-

 Coin A: SD = $30

Coin B: SD = $120

Standard deviation measures volatility — how much the price moves up and down from its average.

Higher SD → Higher volatility → Higher risk

Lower SD → More stable → Lower risk

Coin B is riskier because it has a much higher standard deviation, meaning its price fluctuates more and is less stable.
